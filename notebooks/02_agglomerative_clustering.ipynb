```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from src.agglomerative_clustering import plot_clusters, evaluate_clustering
```

```python
processed_data = pd.read_csv('../data/processed/s1.txt')
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='ward')
labels = agglomerative_clustering.fit_predict(processed_data)
```

```python
plot_clusters(processed_data, labels)
```

```python
evaluate_clustering(processed_data, labels)
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='complete')
labels = agglomerative_clustering.fit_predict(processed_data)
plot_clusters(processed_data, labels)
evaluate_clustering(processed_data, labels)
```
