```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from src.utils_clustering import plot_clusters, evaluate_clustering, load_labels_from_file
```

```python
FILE_NAME = "s1"
file_path = rf'..\data\processed\{FILE_NAME}.txt'
processed_data = pd.read_csv(file_path)
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='ward')
labels_pred = agglomerative_clustering.fit_predict(processed_data)
```

```python
from collections import Counter

cluster_distribution = Counter(labels_pred)
print(cluster_distribution)
```

```python
FILE_NAME = 's1-label'
labels_true = load_labels_from_file(rf'..\data\label\{FILE_NAME}.pa')

cluster_distribution = Counter(labels_true)
print(cluster_distribution)
```

```python
plot_clusters(processed_data, labels_pred, title='Ward Linkage Agglomerative Clustering')
evaluate_clustering(processed_data, labels_pred, labels_pred)
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='complete')
labels_pred = agglomerative_clustering.fit_predict(processed_data)
plot_clusters(processed_data, labels_pred, title='Complete Linkage Agglomerative Clustering')
evaluate_clustering(processed_data, labels_true, labels_pred)
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='average')
labels_pred = agglomerative_clustering.fit_predict(processed_data)
plot_clusters(processed_data, labels_pred, title='Average Linkage Agglomerative Clustering')
evaluate_clustering(processed_data, labels_true, labels_pred)
```

```python
agglomerative_clustering = AgglomerativeClustering(n_clusters=15, linkage='single')
labels = agglomerative_clustering.fit_predict(processed_data)
plot_clusters(processed_data, labels, title='Single Linkage Agglomerative Clustering')
evaluate_clustering(processed_data, labels_true, labels_pred)
```
