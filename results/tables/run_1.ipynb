```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "name": ["KMeans", "DBSCAN", "AffinityPropagation", "Birch", "fcm", "OPTICS", 
             "MeanShift", "cure", "BisectingKMeans", "clique", 
             "AgglomerativeClustering", "rock", "HDBSCAN"],
    "time": [0.000666666, 0.000400039, 0.364424721, 0.000666666, 0.009702635, 
             0.03063494, 0.073140248, 0.014913181, 0.081222196, 0.000483433, 
             0.005169078, 623.575618, 0.002033695],
    "accuracy": [99.38, 96.58, 99.36, 99.44, 91.56, 98.52, 99.38, 99.26, 98.14, 
                 79.32, 99.22, 6.38, 95.26]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
fig, ax1 = plt.subplots(figsize=(14, 8))

# Twin the axis for two different measurements
ax2 = ax1.twinx()

# Plotting time
ax1.bar(df['name'], df['time'], color='g', label='Run Time (seconds)')
ax1.set_xticklabels(df['name'], rotation=90)
ax1.set_xlabel('Clustering Algorithm')
ax1.set_ylabel('Run Time (seconds)', color='g')

# Plotting accuracy
ax2.plot(df['name'], df['accuracy'], color='b', marker='o', label='Accuracy (%)')
ax2.set_ylabel('Accuracy (%)', color='b')

# Title and Legends
plt.title('Clustering Algorithm Comparison: Run Time and Accuracy')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Save the figure
plt.savefig('clustering_algorithm_comparison.png', dpi=300, bbox_inches='tight')

plt.show()
```
